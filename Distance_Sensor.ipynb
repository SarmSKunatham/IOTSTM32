```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [5.63030577 , 11.18000976, 18.3097076, 23.209959, 29.8182354] # actual
y = [5, 10, 15, 20, 25] # metrics
```

```python
plt.figure(figsize=(21, 8))
plt.plot(x, y, 'r+')
plt.xlabel("Measured")
plt.ylabel("Actual")
plt.title("Distance result")
plt.show()
```

```python
x = np.array(x)
y = np.array(y)

m, b = np.polyfit(x, y ,1)

print(f"m: {m}")
print(f"b:  {b}")

plt.figure(figsize=(21, 8))
plt.plot(x, y, 'o')
plt.xlabel("Measured") 
plt.ylabel("Actual")
plt.title("Best fit for ultrasonic sensor \n m: 0.8257889114738466 \n b:  0.4416358795083021")
plt.plot(x, m*x + b)

plt.show()
```
